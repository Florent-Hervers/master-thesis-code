As it's very difficult to infer the best choice of hyperparameters from the evolution of one run. I will take advantage of the previous code that implement bayesian optimisation to determine the best hyperparameters. Some modifications are made in order to optimize on the correlation on the validation set instead of the validation objective (MAE). The bo algorithm will chose the M and learning rate and the hidden_layer size parameters and the code will sweep the different values of lambda like proposed in the original paper.

The hyperparmeters range for the bayesian optimisation are:
```yaml
setting:
  init_random_points: 5

  bo_iterations: 20
```
